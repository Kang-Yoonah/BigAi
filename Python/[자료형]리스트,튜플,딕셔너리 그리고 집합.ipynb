## 2-4 튜플 자료형

튜플(tuple)은 몇 가지 점을 제외하곤 리스트와 거의 비슷하며 리스트와 두드러지게 다른 점은 다음과 같다.
- 리스트는 [ ]으로 둘러싸지만 튜플은 ( )으로 둘러싼다.
- 리스트는 요소 값의 생성, 삭제, 수정이 가능하지만 튜플은 요소 값을 바꿀 수 없다.




````python
t1 = ()
print(t1)
t2 = tuple()
print(t2)
````

````python
# 튜플은 하나의 원소를 가질 땐 반드시 요소 뒤에 콤마(,)를 붙여야한다.
t2 = (1)
print('t2의 타입은 ',type(t2))
t3 = (1,)
print('t3의 타입은 ',type(t3))
````

````python
t5 = ('a','b', ['ab','cd'])
print(t5)
````

튜플과 리스트의 가장 큰 차이는 **요솟값을 변화시킬 수 있는가 여부**이다.   
즉, 리스트의 요솟값은 변화가 가능하고 튜플의 요솟값은 변화가 불가능하다.   
따라서 프로그램이 실행되는 동안 요솟값이 항상 변하지 않기를 바란다면, 튜플을 사용하면 된다.   
이와는 반대로 수시로 그 값을 변화시켜야할 경우라면 리스트를 사용해야한다.

````python
t1 = (1, 2, 'a', 'b')
del t1[0]
````

Unpacking

## 2-5 딕셔너리 자료형

- 키(key)와 값(value)이 한 쌍이 하나의 대응 관계를 가지고 있는 자료형
- 리스트나 튜플처럼 순차적으로 해당 요솟값을 구하지 않고   
딕셔너리는 순서에 상관없이 Key를 통해 Value를 얻는다.
- 따라서 딕셔너리는 순서가 없다.

### 딕셔너리 생성

```
dict = {Key1:Value1, Key2:Value2, Key3:Value3, ...}
```

- Key : 변하지 않는 값
    - 따라서 리스트, 딕셔너리, 집합은 불가능하다.
- Value : 변하는 값과 변하지 않는 값 모두 사용 가능

````python
dic = {'name':'pey', 'phone':'010-9999-1234', 'birth': '1118'}
````

````python
dic
````

````python
# 변수를 통해 할당할 수도
key = "Key"
value = [1,2,3]
a = {key : value}
print(a)
````

### 딕셔너리 쌍 추가, 삭제, 접근, 수정

- `dict[key] = value` 을 통해서 키, 값 쌍을 추가할 수 있다.
- `dict[key]`을 통해서 값에 접근할 수 있다.
    - dict에 해당 key가 없으면 오류 발생
- `dict[key] = new value`을 통해서 값을 변경할 수도 있다.
    - 사실 key를 중복으로 덮어씌우는 것

````python
d1 = {"a" : [1,2,3]}
d2 = {"a" : 1}
````

````python
# 1. 딕셔너리 생성
print("1. 딕셔너리 생성")
d = {'a': 123123}
print(d)
print()
 
# 2. 값 추가
print("2. 값 추가")
d[999] = 10                     # 숫자 키, 숫자 값
d['99'] = 111                   # 문자 키, 숫자 값
d['BlockDMask'] = "blog"        # 문자 키, 문자 값
d['wow'] = [1, 2, 3, 4, 5]      # 문자 키, 리스트 값
d[(1, 2)] = "this is value"     # 튜플 키, 문자 값
d[1] = (3, 'a', 5)              # 숫자 키, 튜플 값
d["dict"] = {"A": a, "B" : b}
print(d)
print()
 
# 3. 값 삭제
print("3. 딕셔너리 값 삭제")
print(f"원래 d : {d}")
del d['dict']
print(f"key 'dict' 삭제 : {d}")
print()

# 4. 값 접근
print("4. 딕셔너리 값 접근")
print(f'd["a"] = {d["a"]}')
print(f'd[999] = {d[999]}')
print(f'd[1] = {d[1]}')        # 숫자 키는 index와 헷갈릴 수 있음
print(f'd[(1,2)] = {d[(1,2)]}')
print(f'd["wow"] = {d["wow"]}')
print(f'd["BlockDMask"] = {d["BlockDMask"]}')
 
# 5. 값 변경
print()
print("5. 딕셔너리 값 변경")
print(f'before : d["BlockDMask"] = {d["BlockDMask"]}')
d["BlockDMask"] = 'boy'
print(f'after  : d["BlockDMask"] = {d["BlockDMask"]}')
 
print(f'before : d["a"] = {d["a"]}')
d["a"] = 'asdf1234'
print(f'after  : d["a"] = {d["a"]}')
````

## 2-6 집합 자료형

- 집합 자료형
    - `set()` 와 같이 생성할 수 있다.
    - 다만 () 안에 값은 순서가 있는 하나의 자료형만 들어갈 수 있어서, list나 문자열, 튜플 등이 들어 갈 수 있다.
- 특징
    - 중복을 허용하지 않는다.
    - 순서가 없다
        - 따라서 인덱스를 가지고 있지 않다.
- 연산
    - `&` or `set1.intersection(set2)` : 교집합
    - `|` or `set1.union(set2)` : 합집합
    - `set1 - set2` or `set1.difference(set2)` : 차집합
- 함수
    - `add()` : 한 개의 값만 추가
    - `update()` : 여러 개의 값 한꺼번에 추가
    - `remove()` : 특정 값 제거
