## 문자열 인덱싱과 슬라이싱

image.png

**인덱싱**
- 인덱싱 : 어떤 것을 "가리킨다"
- 문자열을 이루는 각각의 문자마다 위치를 나타내는 인덱스(index)가 존재한다.
- 인덱스는 0부터 시작한다.
- 음수를 붙여 뒤에서부터 시작할 수 있다.
- `문자열[번호]` 형태로 문자열 안의 특정한 값을 뽑아낼 수 있다.

```python
s = "Hello, world!"
print(s[-6])
print(s[7])
print()
print(s[0], s[-0])
```

**슬라이싱**
- 슬라이싱 : 어떤 것을 "잘라낸다"
- `문자열[시작지점:끝나는 지점+1]` 형태로 연속된 여러 개의 문자를 뽑아낼 수 있다.
- 인덱스 생략도 가능하다. 생략한다면 처음 또는 끝을 의미한다.
- 증가폭도 조정할 수 있다.

```python
s = "Hello, world!"

print("Hello를 가져오려면 어떻게 슬라이싱 해야할까?")
print("방법 1 : ")
print("방법 2 : ")
```

## 문자열 내장 함수

- 문자열 자료형 자체적으로 가지고 있는 함수.
- `문자열.함수이름()` 형식으로 사용한다.
- 종류
    - count() : 문자 개수 세기
    - find() : 위치 알려주기
    - index() : 위치 알려주기2
    - join() : 문자열 삽입
    - upper() : 소문자 -> 대문자
    - lower() : 대문자 -> 소문자
    - strip() : 양쪽 공백 지우기
       - lstrip() : 왼쪽 공백 지우기
       - rstrip() : 오른쪽 공백 지우기
    - replace() : 문자열 바꾸기
    - split() : 문자열 나누기
